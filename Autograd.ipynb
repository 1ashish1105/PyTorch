# Working on a Neural Network to cal student is placed or not :-

```python
x=torch.tensor(6.7) # Input Feature
y=torch.tensor(0) # True Prediction

w=torch.tensor(1)  # Weight
b=torch.tensor(0)  # Bias
```

```python
# Calculating loss (Binary Cross Entropy)\

def binary_cross_entropy_loss(prediction,target):
  epsilon=1e-8 # To prevent log 0.
  prediction = torch.clamp(prediction,epsilon,1-epsilon)
  return -(target * torch.log(prediction)+(1-target)*torch.log(1-prediction))
```

```python
z=w*x+b
y_pred=torch.sigmoid(z)
loss=binary_cross_entropy_loss(y_pred,y)
```

```python
loss
```

```python
# Derivatives:
# 1. dL/d(y_pred): Loss with respect to the prediction (y_pred)
dloss_dy_pred = (y_pred - y)/(y_pred*(1-y_pred))

# 2. dy_pred/dz: Prediction (y_pred) with respect to z (sigmoid derivative)
dy_pred_dz = y_pred * (1 - y_pred)

# 3. dz/dw and dz/db: z with respect to w and b
dz_dw = x  # dz/dw = x
dz_db = 1  # dz/db = 1 (bias contributes directly to z)

dL_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dL_db = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print(f"Manual Gradient of loss w.r.t weight (dw): {dL_dw}")
print(f"Manual Gradient of loss w.r.t bias (db): {dL_db}")
```

## Disable gradient traking(or we want to stop cal derivative):-

```python
x=torch.tensor(2.0,requires_grad=True)
y=x**2
y.backward()
x.grad
```

```python
# 1. Do require grad=False
x.requires_grad=False
```

```python
x
y=x**2
y.backward() #Here backward dont run because we do requires gradient = false.
```

```python
# 2. Using "detach" function we can do this

# Because detach function helps us to create a new silimar tensor
```

```python
x=torch.tensor(2.0,requires_grad=True)
x
z=x.detach()
```

```python
y=x**2
y
# Here we can see that y can do Autograd and it can calulate diffrentiation using grad
```
